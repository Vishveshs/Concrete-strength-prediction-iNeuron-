```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,mean_squared_log_error,max_error
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../input/concrete-compressive-strength-data-set/concrete_data.csv")
df1 = df.copy()
df.head()
```

```python
y = df1['concrete_compressive_strength'].copy()
X = df1.drop('concrete_compressive_strength',axis=1).copy()
```

```python
#X.head()
#y.head()

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.85,random_state=1) 
```

```python
#scaling the features
scaler = StandardScaler()
scaler.fit(X_train)

X_train = pd.DataFrame(scaler.transform(X_train),columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test),columns=X_test.columns)
```

```python
#X_test.head()
#X_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
#models
model_dict = {
    'RandomForestRegressor': RandomForestRegressor(),
    'neural_network': MLPRegressor(),
    'linear_model': LinearRegression(),
    'Decisiontree': DecisionTreeRegressor(),
    'GBR': GradientBoostingRegressor(),
    'svm': SVR()
}
```

```python
#model fitting
for model_name,model in model_dict.items():
    model.fit(X_train,y_train)
    print(f'r squared value for {model_name} is {model.score(X_test,y_test)}')
```

```python
#optimizing the model RandomForest regressor
RFR_model = RandomForestRegressor()

parameters = {
    'n_estimators':[75,100,150,200]
    
    
}
opt_random_forest_model = GridSearchCV(RFR_model,parameters)
opt_random_forest_model.fit(X_train,y_train)
opt_random_forest_model.best_params_
preds = opt_random_forest_model.predict(X_test)
```

```python
print(f'r squared value for RandomForestRegressor is {opt_random_forest_model.score(X_test,y_test)}')
```
